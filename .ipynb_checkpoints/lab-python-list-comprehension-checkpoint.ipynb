````python
inventory = {}
def initialize_inventory(products):
     return {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products} 
    
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = initialize_inventory(products)
inventory
````

````python
count = int(input("Enter the number of customer orders: "))
customer_orders = set()

def get_customer_orders():
    """
        Returns the customer orders set

        Parameters: Empty.
    """
    orders = {input(f"Enter the product {i+1}: ") for i in range(count)}
    return orders

customer_orders = get_customer_orders()
print(customer_orders)

print("Total products ordered: ",count)
percentage_ordered = (count / len(inventory.values())) * 100
print(f"Percentage of Unique Products Ordered : {percentage_ordered:.2f}")
````

````python
updated_inventory = {}
products = []
ordered_list = list(customer_orders)

def update_inventory():
    """
        Returns the updated inventory

        Parameters: customer_orders, inventory.
    """
    for product in customer_orders:
        while True:
            quantity = input(f"Enter the quantity for the product {product}: ")
            if quantity.isdigit():
                quantity = int(quantity)
                break
            print("Please enter a valid numeric quantity.")
       
        products.append(quantity)
        #print(product)
        inventory_quantity = inventory.get(product)
        #print(quantity)
        # Update inventory
        if inventory_quantity is None:
            print(f"{product} not in inventory, skipping.")
            continue
        new_qty = inventory_quantity - int(quantity)
        if(new_qty <= 0):
            inventory.pop(product)
        else:
            inventory[product] = new_qty
    
update_inventory()
print('updated Inventory:')

[print(f"{product}: {quantity}") for product, quantity in inventory.items()]
print(f"Inventory: ",{p: q for p, q in inventory.items() if q > 0})
````

````python
prodcuct_items = []
total_price = float()

#function to calculate the total price of the customer order.
def totaPriceOrder():
    """
        Calculate the total price of the customer order

        Parameters: Empty.
    """
    for product in customer_orders:
        price = float(input(f"Enter the price of the product {product}: "))
            
        prodcuct_items.append(float(price))
    price = sum(float(price) for price in prodcuct_items)
    return price
    
total_price = totaPriceOrder()
print(f"Total price: {total_price}")
````
